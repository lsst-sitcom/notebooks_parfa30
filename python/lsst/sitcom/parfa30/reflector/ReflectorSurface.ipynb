```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_parker = pd.read_csv('reflector_surface_20230217_parker.csv',skiprows=[0,2])
df_brian = pd.read_csv('reflector_surface_20230217_brian.csv',skiprows=[0,2])
df_old = pd.read_csv('reflector_surface_201808.csv',skiprows=[0,2])
```

```python
plt.scatter(df_parker.X2, df_parker.Y2,c=df_parker.Mag)
plt.scatter(df_brian.X2, df_brian.Y2,c=df_brian.Mag)
plt.colorbar()
```

```python
plt.scatter(df_old.X2, df_old.Y2,c=df_old.Mag)
plt.colorbar()
```

```python
df_parker['r2'] = np.sqrt(df_parker['X2']**2 + df_parker['Y2']**2)
plt.plot(df_parker.r2, df_parker.Mag, 'x',label='Feb 2023')
```

```python
df_brian['r2'] = np.sqrt(df_brian['X2']**2 + df_brian['Y2']**2)
df_brian['r1'] = np.sqrt(df_brian['X1']**2 + df_brian['Y1']**2)
df_old['r1'] = np.sqrt(df_old['X2']**2 + df_old['Y2']**2)
```

```python
fig, ax = plt.subplots(1, figsize=(10,4))
ax2 = ax.twiny()
ax3 = ax.twinx()
ax.plot(df_brian.r1, df_brian.Mag, 'x',label='Feb 2023 - brian',c='C0')
ax3.plot(df_old.r1*25.4, df_old.Mag, 'x',label='Aug 2018',c='C1')
#ax2.plot(df_parker.r2, df_parker.Mag, 'x',label='Feb 2023 - parker',c='C2')
ax.set_xlabel('Radius (mm)')
ax.legend()
ax2.legend(loc='upper left')
ax3.legend(loc='lower left')
ax.set_ylabel('Offset Feb 2023 (mm)')
ax3.set_ylabel('Offset Aug 2018 (mm)')
#plt.ylabel('Offset (mm)')
#plt.title('Laser Tracker Measurement of Reflector Feb. 15, 2023')
```
